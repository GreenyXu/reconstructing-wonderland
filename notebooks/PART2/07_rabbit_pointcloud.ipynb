```python
import os
import numpy as np
import trimesh
import pandas as pd

# =========================
# 参数设置
# =========================
INPUT_DIR = "../../data/raw/rabbit_models"
OUTPUT_DIR = "../../data/processed/rabbit_pointclouds"
N_POINTS = 5000

os.makedirs(OUTPUT_DIR, exist_ok=True)

# =========================
# 处理函数
# =========================
def mesh_to_pointcloud(mesh_path, n_points=5000):
    mesh = trimesh.load(mesh_path)

    # 确保是mesh
    if isinstance(mesh, trimesh.Scene):
        mesh = mesh.dump(concatenate=True)

    # 表面采样
    points, _ = trimesh.sample.sample_surface(mesh, n_points)

    return points

# =========================
# 批处理
# =========================
files = [
    "rabbit_mesh_01_slim.obj",
    "rabbit_mesh_02_standard.obj",
    "rabbit_mesh_03_fat.obj"
]

for file in files:
    input_path = os.path.join(INPUT_DIR, file)

    print(f"Processing: {file}")

    pts = mesh_to_pointcloud(input_path, N_POINTS)

    df = pd.DataFrame(pts, columns=["x", "y", "z"])

    output_name = file.replace(".obj", ".csv")
    output_path = os.path.join(OUTPUT_DIR, output_name)

    df.to_csv(output_path, index=False)

    print(f"Saved: {output_path}")

print("Done!")
```
