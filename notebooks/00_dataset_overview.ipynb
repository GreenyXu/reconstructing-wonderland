# 00 — Dataset Overview and Project Framing

This notebook defines the project theme, introduces the datasets, and outlines the cross-domain workflow.

Project theme:
**Translating literary spatial imagination into architectural form through text, images, and machine learning.**

Primary literary source:
**Italo Calvino, *Invisible Cities***

## Project framing

This project explores how literary descriptions of cities can be translated into spatial and architectural form through a data-driven workflow.

The core idea is to build a cross-domain pipeline:

**literary text → structured text units → visual references → embeddings / machine learning → spatial mapping → 3D form generation**

Rather than treating literature as a purely interpretive source, the project uses computational methods to extract patterns, compare domains, and define transformation rules for later design outputs.

## Datasets

The project is organised around three datasets:

### Dataset 1 — Literary text
Source: *Invisible Cities* by Italo Calvino  
Type: literary text units  
Role: provides spatial imagination, descriptive language, and narrative cues

### Dataset 2 — Curated Flickr image dataset
Source: Flickr API  
Type: urban / architectural photographs organised by city-type labels  
Role: provides a visual spatial corpus for cross-domain comparison

### Dataset 3 — Derived image-text dataset
Source: generated from the Flickr corpus through captioning / image analysis  
Type: captions, tags, emotion cues, and later embeddings  
Role: bridges visual data and textual description for machine learning comparison

These datasets will later be vectorised and connected through retrieval, clustering, and spatial mapping.

## Workflow overview

The full workflow of the project is structured into the following stages:

### Part 1 — Projecting between domains
1. Process literary text into structured units
2. Build / curate a Flickr-based urban image dataset
3. Retrieve image references for each literary unit
4. Vectorise text and images
5. Compare representations and cluster patterns

### Part 2 — Translating data into spatial form
6. Convert analytical outputs into spatial rules
7. Generate architectural fragments
8. Recompose fragments into scenes and final visual outputs

This means the project is not simply about collection or analysis:  
it aims to connect all stages into a single data-driven design workflow.
