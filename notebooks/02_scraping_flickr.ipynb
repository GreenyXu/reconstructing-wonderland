# 02 - Flickr scraping test

## Aim
This notebook collects publicly visible Flickr image pages related to brutalist architecture and extracts basic metadata for testing.

## Keywords
- brutalist architecture
- concrete building
- brutalism facade

## Outputs
- image page URLs
- image URLs
- titles
- tags
- metadata CSV

```python
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin, quote
import pandas as pd
import time
```

```python
HEADERS = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36",
    "Accept-Language": "en-GB,en;q=0.9"
}
```

```python
keyword = "brutalist architecture"
search_url = f"https://www.flickr.com/search/?text={quote(keyword)}"

resp = requests.get(search_url, headers=HEADERS, timeout=20)
print(resp.status_code)
print(resp.url)
print(resp.text[:500])
```

```python
soup = BeautifulSoup(resp.text, "lxml")

all_links = []
for a in soup.find_all("a", href=True):
    href = a["href"]
    full = urljoin("https://www.flickr.com", href)
    all_links.append(full)

unique_links = sorted(set(all_links))
print("Total unique links:", len(unique_links))
for link in unique_links[:100]:
    print(link)
```
