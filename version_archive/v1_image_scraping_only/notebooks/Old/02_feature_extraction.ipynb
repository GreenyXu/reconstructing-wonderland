# 03 - Image Vectorisation: CLIP and HOG

## Aim
This notebook compares two image vectorisation methods for the Flickr brutalist architecture dataset:

- CLIP embeddings for semantic representation
- HOG features for structural representation

The goal is to understand how different visual representations can be captured in different feature spaces and later translated into spatial surface generation rules.

```python
from pathlib import Path
import pandas as pd

resize_dir = Path("../data/images_resized")
image_paths = sorted([str(p) for p in resize_dir.glob("*.jpg")])

print(len(image_paths))
print(image_paths[:5])
```

```python
from PIL import Image
import numpy as np
from skimage.feature import hog
from skimage.color import rgb2gray

hog_features = []
hog_paths = []

for path in image_paths:
    try:
        img = Image.open(path).convert("RGB").resize((256, 256))
        img_np = np.array(img)
        gray = rgb2gray(img_np)

        feat = hog(
            gray,
            orientations=9,
            pixels_per_cell=(16, 16),
            cells_per_block=(2, 2),
            block_norm="L2-Hys"
        )

        hog_features.append(feat)
        hog_paths.append(path)
    except Exception as e:
        print("HOG failed:", path, e)

hog_features = np.array(hog_features)
print(hog_features.shape)
```

```python
import open_clip
import torch
from PIL import Image
import numpy as np

device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model, _, preprocess = open_clip.create_model_and_transforms(
    "ViT-B-32",
    pretrained="openai"
)
model = model.to(device)
model.eval()
```

```python
clip_embeddings = []
clip_paths = []

for path in image_paths:
    try:
        image = Image.open(path).convert("RGB")
        image_tensor = preprocess(image).unsqueeze(0).to(device)

        with torch.no_grad():
            features = model.encode_image(image_tensor)
            features = features / features.norm(dim=-1, keepdim=True)

        clip_embeddings.append(features.cpu().numpy().flatten())
        clip_paths.append(path)

    except Exception as e:
        print("CLIP failed:", path, e)

clip_embeddings = np.array(clip_embeddings)
print(clip_embeddings.shape)
```
