# 01 Project Setup

## Goal
This notebook sets up the project structure for the new multimodal version of **data-driven-surface**.

The project integrates four data sources:
1. literary text
2. scraped images
3. AI-generated text
4. image-derived data

This notebook will:
- define the project structure
- create key folders if needed
- check the Python environment
- save reusable project paths

## Step 1 — Import libraries
Import the basic Python libraries used for path management, data handling, and environment checks.

```python
from pathlib import Path
import json
import sys
import platform
import pandas as pd
```

## Step 2 — Define the project root
Set the project root folder and create reusable path variables for all major directories.

```python
# Set project root
PROJECT_ROOT = Path.cwd().resolve().parent

DATA_DIR = PROJECT_ROOT / "data"
RAW_DIR = DATA_DIR / "raw"
CLEANED_DIR = DATA_DIR / "cleaned"
PROCESSED_DIR = DATA_DIR / "processed"

NOTEBOOKS_DIR = PROJECT_ROOT / "notebooks"
OUTPUTS_DIR = PROJECT_ROOT / "outputs"
SRC_DIR = PROJECT_ROOT / "src"
FIGURES_DIR = PROJECT_ROOT / "figures"

# Raw data sources
RAW_LITERARY_DIR = RAW_DIR / "literary_text"
RAW_IMAGES_DIR = RAW_DIR / "scraped_images"
RAW_AI_TEXT_DIR = RAW_DIR / "ai_generated_text"
RAW_IMAGE_DERIVED_DIR = RAW_DIR / "image_derived_data"

# Cleaned data sources
CLEANED_LITERARY_DIR = CLEANED_DIR / "literary_text"
CLEANED_IMAGES_DIR = CLEANED_DIR / "scraped_images"
CLEANED_AI_TEXT_DIR = CLEANED_DIR / "ai_generated_text"
CLEANED_IMAGE_DERIVED_DIR = CLEANED_DIR / "image_derived_data"

# Processed data sources
PROCESSED_LITERARY_DIR = PROCESSED_DIR / "literary_text"
PROCESSED_IMAGES_DIR = PROCESSED_DIR / "scraped_images"
PROCESSED_AI_TEXT_DIR = PROCESSED_DIR / "ai_generated_text"
PROCESSED_IMAGE_DERIVED_DIR = PROCESSED_DIR / "image_derived_data"

# Output folders
OUTPUT_FIGURES_DIR = OUTPUTS_DIR / "figures"
OUTPUT_TABLES_DIR = OUTPUTS_DIR / "tables"
OUTPUT_MAPPINGS_DIR = OUTPUTS_DIR / "mappings"
OUTPUT_PROMPTS_DIR = OUTPUTS_DIR / "prompts"
OUTPUT_GENERATED_IMAGES_DIR = OUTPUTS_DIR / "generated_images"
```

## Step 3 — Create folders if they do not exist
This ensures that the notebook can run safely even if some folders are missing.

```python
folders_to_check = [
    DATA_DIR,
    RAW_DIR, CLEANED_DIR, PROCESSED_DIR,
    RAW_LITERARY_DIR, RAW_IMAGES_DIR, RAW_AI_TEXT_DIR, RAW_IMAGE_DERIVED_DIR,
    CLEANED_LITERARY_DIR, CLEANED_IMAGES_DIR, CLEANED_AI_TEXT_DIR, CLEANED_IMAGE_DERIVED_DIR,
    PROCESSED_LITERARY_DIR, PROCESSED_IMAGES_DIR, PROCESSED_AI_TEXT_DIR, PROCESSED_IMAGE_DERIVED_DIR,
    NOTEBOOKS_DIR, OUTPUTS_DIR, SRC_DIR, FIGURES_DIR,
    OUTPUT_FIGURES_DIR, OUTPUT_TABLES_DIR, OUTPUT_MAPPINGS_DIR, OUTPUT_PROMPTS_DIR, OUTPUT_GENERATED_IMAGES_DIR
]

for folder in folders_to_check:
    folder.mkdir(parents=True, exist_ok=True)

print("Project folders checked and created if necessary.")
```

## Step 4 — Display the main project paths
Print the current project structure to confirm that all important folders are correctly linked.

```python
project_paths = {
    "PROJECT_ROOT": str(PROJECT_ROOT),
    "RAW_LITERARY_DIR": str(RAW_LITERARY_DIR),
    "RAW_IMAGES_DIR": str(RAW_IMAGES_DIR),
    "RAW_AI_TEXT_DIR": str(RAW_AI_TEXT_DIR),
    "RAW_IMAGE_DERIVED_DIR": str(RAW_IMAGE_DERIVED_DIR),
    "CLEANED_LITERARY_DIR": str(CLEANED_LITERARY_DIR),
    "PROCESSED_LITERARY_DIR": str(PROCESSED_LITERARY_DIR),
    "OUTPUT_PROMPTS_DIR": str(OUTPUT_PROMPTS_DIR),
    "OUTPUT_GENERATED_IMAGES_DIR": str(OUTPUT_GENERATED_IMAGES_DIR)
}

for key, value in project_paths.items():
    print(f"{key}: {value}")
```

## Step 5 — Check the Python environment
This step records the Python version and operating system for reproducibility.

```python
print("Python version:", sys.version)
print("Platform:", platform.platform())
print("Pandas version:", pd.__version__)
```

## Step 6 — Save project paths as a JSON config file
This file can be reused by later notebooks to avoid redefining paths every time.

```python
config = {
    "project_root": str(PROJECT_ROOT),
    "data_dir": str(DATA_DIR),
    "raw_literary_dir": str(RAW_LITERARY_DIR),
    "raw_images_dir": str(RAW_IMAGES_DIR),
    "raw_ai_text_dir": str(RAW_AI_TEXT_DIR),
    "raw_image_derived_dir": str(RAW_IMAGE_DERIVED_DIR),
    "cleaned_literary_dir": str(CLEANED_LITERARY_DIR),
    "processed_literary_dir": str(PROCESSED_LITERARY_DIR),
    "outputs_dir": str(OUTPUTS_DIR),
    "output_figures_dir": str(OUTPUT_FIGURES_DIR),
    "output_tables_dir": str(OUTPUT_TABLES_DIR),
    "output_mappings_dir": str(OUTPUT_MAPPINGS_DIR),
    "output_prompts_dir": str(OUTPUT_PROMPTS_DIR),
    "output_generated_images_dir": str(OUTPUT_GENERATED_IMAGES_DIR),
    "src_dir": str(SRC_DIR),
    "processed_image_derived_dir": str(PROCESSED_IMAGE_DERIVED_DIR)
}

config_path = PROJECT_ROOT / "src" / "project_config.json"

with open(config_path, "w", encoding="utf-8") as f:
    json.dump(config, f, indent=4, ensure_ascii=False)

print(f"Config saved to: {config_path}")
```

## Step 7 — Preview the config file
Load the saved config file and check its contents.

```python
with open(config_path, "r", encoding="utf-8") as f:
    loaded_config = json.load(f)

loaded_config
```
