## Step 5 — Select the EPUB source file
This notebook uses the EPUB file as the main source because it preserves text structure more reliably than PDF.

```python
if len(epub_files) == 0:
    raise FileNotFoundError(
        "No EPUB file was found in references/literary_sources/. "
        "Please place the Invisible Cities EPUB file there first."
    )

source_epub_path = epub_files[0]
print("Selected EPUB file:", source_epub_path.name)
```

## Step 6 — Read the EPUB file
Load the EPUB book into Python.

```python
book = epub.read_epub(str(source_epub_path))
print("EPUB loaded successfully.")
```

## Step 7 — Extract text from EPUB document items
Each document item is parsed with BeautifulSoup and converted to plain text.

```python
raw_sections = []

for item in book.get_items():
    if item.get_type() == ITEM_DOCUMENT:
        soup = BeautifulSoup(item.get_content(), "html.parser")
        text = soup.get_text(separator="\n")
        raw_sections.append(text)

print("Number of extracted sections:", len(raw_sections))
```

## Step 10 — Define text cleaning functions
Create helper functions for light cleaning while preserving paragraph structure.

The aim here is not full literary analysis yet. This notebook only prepares a readable `.txt` source.

```python
def basic_clean_text(text: str) -> str:
    """
    Basic text cleaning while preserving paragraph breaks.
    """
    text = text.replace("\r\n", "\n").replace("\r", "\n")
    text = text.replace("\xa0", " ")

    # Normalize unicode ellipsis
    text = text.replace("…", "...")

    # Remove excessive blank lines
    text = re.sub(r"\n{3,}", "\n\n", text)

    # Remove repeated spaces or tabs
    text = re.sub(r"[ \t]+", " ", text)

    # Remove spaces at line ends
    text = re.sub(r" +\n", "\n", text)

    return text.strip()


def remove_common_front_matter(text: str) -> str:
    """
    Remove front matter, TOC noise, page-number-like residues,
    and fragmented contents lines.
    """
    # Remove pure number lines
    text = re.sub(r"(?m)^\s*\d+\s*$", "", text)

    # Remove lines like "20 ...."
    text = re.sub(r"(?m)^\s*\d+\s*\.{2,}\s*$", "", text)

    # Remove single dash lines
    text = re.sub(r"(?m)^\s*[-–—]\s*$", "", text)

    # Remove TOC lines with trailing numbers
    # e.g. "15 Cities and memory 4"
    text = re.sub(r"(?mi)^\s*\d+\s+[A-Za-z][A-Za-z\s&]+\s+\d+\s*$", "", text)

    # Remove fragmented TOC title lines
    # e.g. "Cities and memory", "Thin cities", "Trading cities"
    text = re.sub(
        r"(?mi)^\s*(cities and [a-z]+|thin cities|trading cities)\s*$",
        "",
        text
    )

    lines = text.split("\n")
    filtered_lines = []

    skip_patterns = [
        r"^contents$",
        r"^table of contents$",
        r"^copyright$",
        r"^title page$",
        r"^about the author$",
        r"^first published.*$",
        r"^translated from.*$",
        r"^italo calvino$",
        r"^invisible cities$"
    ]

    for line in lines:
        line_clean = line.strip().lower()
        if not line_clean:
            filtered_lines.append("")
            continue
        if any(re.match(pattern, line_clean) for pattern in skip_patterns):
            continue
        filtered_lines.append(line)

    text = "\n".join(filtered_lines)

    # Remove remaining author-bio tail
    text = re.sub(r"(?i)\bthe Premio Feltrinelli\.", "", text)

    # Remove stray dash lines
    text = re.sub(r"\n\s*[-–—]\s*\n", "\n", text)

    # Clean up blank lines again after removals
    text = re.sub(r"\n{3,}", "\n\n", text)

    return text.strip()


def transform_city_headers(text: str) -> str:
    """
    Convert genuine chapter headers like:
    CITIES & MEMORY 1
    into:
    [CITY_GROUP: Cities & Memory 1]
    """
    def replace(match):
        header = match.group(0).strip()
        return f"\n[CITY_GROUP: {header.title()}]\n"

    return re.sub(
        r"(?mi)^\s*(cities\s*&\s*[a-z\s]+?\d+|thin\s+cities\s+\d+|trading\s+cities\s+\d+)\s*$",
        replace,
        text
    )
```

## Step 11 — Apply initial cleaning
Run the light cleaning pipeline on the extracted EPUB text.

```python
cleaned_text = basic_clean_text(raw_combined_text)
cleaned_text = remove_common_front_matter(cleaned_text)
cleaned_text = transform_city_headers(cleaned_text)

print(cleaned_text[:4000])
```
